### Part a:
Use a Runge–Kutta algorithm such as RK4 or RK45 (see Appendix 3A) to obtain a numerical solution of the Lorenz equations (6.33). Generate three-dimensional plots using Display3DFrame. Explore the basin of the attractor with $σ = 10$, $b = \frac{8}{3}$, and $r = 28$

```python
rk = om.RungeKutta4()
lorenz = l.Lorenz()
integrator = om.Integrator(ode=lorenz, method=rk)

t0 = [0, 100]
dt = 0.01
u0 = [0.0, 1.0, 10.0]

t, u = integrator.integrate(interval=t0, u_0=u0, dt=dt)

ax = plt.figure().add_subplot(projection='3d')

ax.plot(*u.T, linewidth=0.25)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```

### Part c:
Let $z_m$ denote the value of $z$ where $z$ is a relative maximum for the $m\text{th}$ time. You can
determine the value of $z_m$ by finding the average of the two values of $z$ when the right-hand
side of (6.33) changes sign. Plot $z_{m+1}$ versus $z_m$ and describe what you find. This procedure
is one way that a continuous system can be mapped onto a discrete map. What is the slope
of the $z_{m+1}$ versus $z_m$ curve? Is its magnitude always greater than unity? If so, then this
behavior is an indication of chaos. Why?

```python
lorenz = l.Lorenz()
rk = om.RungeKutta4()
integrator = om.Integrator(ode=lorenz, method=rk)

t0 = [0, 100]
dt = 0.01
u0 = [0.0, 1.0, 10.0]

t, u = integrator.integrate(interval=t0, u_0=u0, dt=dt)

z_maxima = []
z_maxima_only = []
for i in range(1, len(u) - 1):
    if u[i-1][2] < u[i][2] and u[i][2] > u[i+1][2]:
        z_maxima.append(u[i])
        z_maxima_only.append(u[i][2])

z_maxima = np.array(z_maxima)
z_maxima_only = np.array(z_maxima_only)

# Plotting z_{m+1} versus z_m
plt.figure()
plt.plot(z_maxima_only[:-1], z_maxima_only[1:], 'o', markersize=3)
plt.xlabel("$z_m$")
plt.ylabel("$z_{m+1}$")
plt.title("$z_{m+1}$ vs $z_m$")
plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.plot(*u.T, linewidth=0.25)
ax.plot(*z_maxima.T, 'ro', markersize=3)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor with Maximum z Points")

plt.show()
```
